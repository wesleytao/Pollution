```python
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
#word cloud without stopping words
stopwords_set = set(stopwords.words("english"))
cleantext = []
for sentence in nltk.sent_tokenize(sentences):
    print sentence
    words_filtered = nltk.word_tokenize(sentence) 
    words_filtered = [WNlemma.lemmatize(t.lower()) for t in words_filtered]
    words_without_stopwords = [word for word in words_filtered if not word in stopwords_set]
    cleantext.extend(words_without_stopwords)

wordlist = nltk.FreqDist(cleantext)
features = wordlist.keys()

print 
print features
```

```python
print 'clean wordcloud'
wordcloud_draw(features,'white')
```

```python
#Feature Extraction
```

```python
# Read in the data
df = pd.read_csv('Amazon_Unlocked_Mobile.csv')
df = df.sample(frac=0.1, random_state=10)
df.dropna(inplace=True)
df = df[df['Rating'] != 3]

df['Positively Rated'] = np.where(df['Rating'] > 3, 1, 0)
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Reviews'], 
                                                    df['Positively Rated'], 
                                                    random_state=0)
```

```python
print X_train.shape
print
print X_train.head()
```

```python

vect = CountVectorizer().fit(X_train)

X_train_vectorized = vect.transform(X_train)

feature_names = np.array(vect.get_feature_names())

print X_train.shape
print X_train_vectorized.shape #document * volabulary

print feature_names[-10:]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

sorted_coef_index = model.coef_[0].argsort()
print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect = TfidfVectorizer(min_df=5).fit(X_train)
X_train_vectorized = vect.transform(X_train)

feature_names = np.array(vect.get_feature_names())

print X_train.shape
print X_train_vectorized.shape #document * volabulary

print feature_names[-10:]
```

```python
#N-gram
vect = CountVectorizer(min_df=5, ngram_range=(1,3), stop_words='english').fit(X_train)
X_train_vectorized = vect.transform(X_train)

feature_names = np.array(vect.get_feature_names())

print X_train.shape
print X_train_vectorized.shape #document * volabulary

print feature_names[-10:]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import math

from nltk.sentiment.vader import SentimentIntensityAnalyzer

example = "This is an example sentence! However, it " \
          "is a very informative one,"

print nltk.word_tokenize(example, language='english')

tokenizer = RegexpTokenizer(r'\w+')
print tokenizer.tokenize(example)
```

```python
stop_words = set(stopwords.words('english'))

all_words = []
for sentence in nltk.sent_tokenize(example):
    print sentence
    words = tokenizer.tokenize(sentence)
    for w in words:
        if w.lower() not in stop_words:
            all_words.append(w.lower())
            
freqDist = nltk.FreqDist(all_words)
print freqDist.keys()[:5]
```

```python
hotel_rev = ["Great place to be when you are in Bangalore.",
"The place was being renovated when I visited so the seating was limited.",
"Loved the ambience, loved the food"]

sid = SentimentIntensityAnalyzer()
for sentence in hotel_rev:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in ss:
        print '{0}: {1}, '.format(k, ss[k])
    print
```
